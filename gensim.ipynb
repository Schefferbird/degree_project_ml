## Gensim Summarization

### Supposed to be an implementation of summa textrank

```python
#!pip install gensim
import gensim
import nltk
from nltk.tokenize import sent_tokenize
```

```python
def gensim_sum(text):
    gensim_summary = gensim.summarization.summarize(text, ratio=0.4)
    summary = nltk.sent_tokenize(gensim_summary)
    
    return ' '.join(summary)
```
