```python
#!python -m spacy download en
import spacy
```

```python
# Save model
nlp = spacy.load('en')

# Save text with spaCy model
spacy_nlp = nlp(text)
```

```python
# Takes a text saved with spaCy model
def spacy_ner(nlp):

    ner_tags = []

    # Save only NER tags
    for ent in nlp.ents:
        ner_tags.append(tuple((ent.text, ent.label_)))
        
    return ner_tags
```

```python
print(spacy_nlp.ents)
```

```python
ner_tagged = spacy_ner(spacy_nlp)

pd.DataFrame(ner_tagged, columns=['Words', 'NE'])[:20]
```

### Displacy NER

```python
from spacy import displacy

# Show ne's in text with displacy
displacy.render(spacy_nlp, jupyter=True, style='ent')
```

### SpaCy Dependency pattern

```python
dependency_pattern = '{left}<---{word}[{w_type}]--->{right}\n--------'

for token in spacy_nlp:
    print(dependency_pattern.format(word=token.orth_, 
                                   w_type=token.dep_,
                                   left=[t.orth_ 
                                            for t 
                                            in token.lefts],
                                   right=[t.orth_ 
                                             for t 
                                             in token.rights]))
```

### SpaCy POS

```python
pos_tagged = [(word, word.tag_, word.pos_) for word in spacy_nlp]
pd.DataFrame(pos_tagged, columns=['Word', 'POS tag', 'Tag type'])[:10]
```
